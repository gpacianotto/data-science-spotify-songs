```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
df = pd.read_csv("../dataset.csv")
# Step 1: Filter the DataFrame by track_genre
desired_genre = 'pop'  # Example track_genre
filtered_df = df[df['track_genre'] == desired_genre]

# Step 2: Create a boxplot
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
filtered_df.boxplot(column=['acousticness'])
plt.title('Acousticness for {}'.format(desired_genre))
plt.ylabel('Acousticness')

plt.subplot(2, 1, 2)
filtered_df.boxplot(column=['duration_ms'])
plt.title('Duration for {}'.format(desired_genre))
plt.ylabel('Duration (ms)')

plt.tight_layout()
plt.show()
```
